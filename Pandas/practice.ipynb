```python
import numpy as np 
import pandas as pd 
```

```python
# Create a Pandas Series from a list.
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
s
```

```python
# Create a Series with custom indexes. 
data = [10, 20, 30, 40, 50]
s = pd.Series((data), index = ['A', 'B', 'C', 'D', 'E'])
s 
```

```python
# Create a Series from a dictionary. 
data = {
    "Math": 90,
    "Science": 85,
    "English": 88
}
s = pd.Series(data)
s
```

```python
# Print the first and last element of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("First Element:", s.iloc[0])
print("Last Element :", s.iloc[-1])
```

```python
# Find the dtype, size, shape, and index of a Series.  
s = pd.Series([10, 20, 30, 40, 50])

print("Data Type:", s.dtype)
print("Size     :", s.size)
print("Shape    :", s.shape)
print("Index    :", s.index)
```

```python
# Calculate the sum, mean, maximum, and minimum of a Series. 
s = pd.Series([10, 20, 30, 40, 50])

print("Sum     :", s.sum())
print("Mean    :", s.mean())
print("Maximum :", s.max())
print("Minimum :", s.min())
```

```python
# Sort a Series by values. 
s = pd.Series([45, 10, 80, 25, 60])

print(s.sort_values()) 
```

```python
# Sort a Series by index.  
s = pd.Series([85, 90, 78],
              index=['C', 'A', 'B'])
print(s.sort_index()) 
```

```python
# Create a DataFrame from a dictionary.  
data = {
    "Name": ["John", "Alice", "Bob"],
    "Age": [25, 30, 28],
    "City": ["New York", "London", "Paris"]
}
df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame from a list of lists.  
data = [
    ["Ram", 21, 85],
    ["Shyam", 22, 90],
    ["Mohan", 20, 78]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])
df
```

```python
# Create a DataFrame from a list of dictionaries.  
data = [
    {"Name": "Ram", "Age": 21, "Marks": 85},
    {"Name": "Shyam", "Age": 22, "Marks": 90},
    {"Name": "Mohan", "Age": 20, "Marks": 78}
]
df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame with custom indexes. 
data = [
    ["Ram", 21, 85],
    ["Shyam", 22, 90],
    ["Mohan", 20, 78]
]
df = pd.DataFrame(data, columns=["Name", "Age", "Marks"], index=["A", "B", "C"])
df
```

```python
# Print the first 5 rows using head().  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Marks": [85, 90, 78, 88, 75, 95]
} 
df = pd.DataFrame(data)
print(df.head()) 
```

```python
# Print the last 3 rows using tail().  
data = {
    "Name": ["Ram", "Shyam", "Mohan", "Amit", "Ravi", "Karan"],
    "Marks": [85, 90, 78, 88, 75, 95]
} 
df = pd.DataFrame(data)
print(df.tail(3))  
```

```python
# Display dataset information using info()  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}
df = pd.DataFrame(data)
df.info()
```

```python
# Display statistical summary using describe()  
data = {
    "Age": [21, 22, 20, 23, 24],
    "Marks": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)
print(df.describe())
```

```python
# Find the shape of a DataFrame.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("Shape:", df.shape)
```

```python
# Find the column names.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("column names:", df.columns)
```

```python
# Find the row indexes.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("row indexes:", df.index)
```

```python
# Find the data types of all columns.  

data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("data types of all columns is:", df.dtypes)
```

```python
# Find the total number of elements.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("total number of elements is:", df.size)
```

```python
# Convert a DataFrame into a NumPy array using values.  
data = {
    "Name": ["Ram", "Shyam", "Mohan"],
    "Age": [21, 22, 20],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("DataFrame into a NumPy array using values:", df.values)
```
