```python
import numpy as np 
import pandas as pd 
```

```python
# Finding Missing Data
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # Null value find
```

```python
df.isna().sum() # Sum 
```

```python
df.isna().any() 
```

```python
df
```

```python
# Removing Missing Data
df.dropna() # Row remove based
```

```python
df
```

```python
df.dropna(thresh=1) # In rows find non null values
```

```python
df
```

```python
# Filling the missing Data 
df.fillna(0)
```

```python
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
