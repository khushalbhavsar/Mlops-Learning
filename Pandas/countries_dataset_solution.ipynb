**Level 1: Data Exploration**

```python
# Load the Countries.csv dataset into a Pandas DataFrame.
import pandas as pd

df = pd.read_csv("Countries.csv") 
```

```python
# Display the first 10 rows and the last 10 rows of the dataset. 
print(df.head(10))

print(df.tail(10))
```

```python
# Find:
# Number of rows and columns
print("Number of rows and columns: ", df.shape)
```

```python
# Column names
print("Column names: ", df.columns)
```

```python
# Data types of all columns
print("Data types of all columns : ", df.dtypes)
```

```python
# Memory usage of the dataset 
print("Memory usage of the dataset : ", df.info(memory_usage="deep"))
```

**Level 2: Selecting Data**

```python
# Display only the following columns:
# country
# capital_city
# continent
# population

print(df[["country","capital_city","continent","population"]]) 
```

```python
# Display the first 15 countries using iloc.  
print(df.iloc[:15]) 
```

```python
# Display the country, GDP, and population of the country at index 50 using loc.  
print(df.loc[50, ["country","gdp","population"]]) 
```

**Level 3: Filtering** 

```python
# Display all countries from Asia.  
print(df[df["continent"]=="Asia"]) 
```

```python
# Display all countries whose population is greater than 100 million.  
print(df[df["population"]>100000000]) 
```

```python
# Internet >80 and Life Expectancy >75  
print(df[(df["internet_pct"]>80) &
         (df["life_expectancy"]>75)]) 
```

```python
# Display all countries having GDP greater than the average GDP.  
avg_gdp = df["gdp"].mean()

print(df[df["gdp"]>avg_gdp]) 
```

**Level 4: Sorting**

```python
# Sort by GDP (Descending) 
print(df.sort_values(by="gdp",ascending=False)) 
```

```python
# Top 10 countries by population 
print(df.sort_values(by="population",
                     ascending=False).head(10)) 
```

```python
# Top 10 highest life expectancy  
print(df.sort_values(by="life_expectancy",
                     ascending=False).head(10)) 
```

**Level 5: Missing Data**

```python
# Missing values in each column  
print(df.isnull().sum()) 
```

```python
# Rows containing missing values  
print(df[df.isnull().any(axis=1)]) 
```

**Level 6: Data Operations**

```python
# Population Density  
df["population_density"] = (
    df["population"] /
    df["land_area"]
)

print(df[["country","population_density"]]) 
```

```python
# Male Percentage  
df["male_percentage"] = (
    df["population_male"] /
    df["population"]
)*100

print(df[["country","male_percentage"]]) 
```

**Level 7: GroupBy**

```python
# Average GDP of each continent  
print(df.groupby("continent")["gdp"].mean()) 
```

```python
# Find
# Maximum population
# Minimum population
# Average life expectancy 

print(
df.groupby("continent").agg({
    "population":["max","min"],
    "life_expectancy":"mean"
})
) 
```

```python
# Count countries in each democracy type  
print(df["democracy_type"].value_counts()) 
```

**Level 8: Pivot Tables**

```python
# Population by continent  
print(
df.pivot_table(
    values="population",
    index="continent",
    aggfunc="sum"
)
) 
```

```python
# Mean GDP by democracy type and continent  
print(
df.pivot_table(
    values="gdp",
    index="democracy_type",
    columns="continent",
    aggfunc="mean"
)
) 
```

**Level 9: Advanced Analysis**
