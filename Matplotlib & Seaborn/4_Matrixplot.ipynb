```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

# Heat map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tipscorr  = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

**Heat Map: A Heatmap is a graphical representation of data where colors represent numerical values.**

```python
sns.heatmap(tipscorr.corr(),annot= True)
```

**Cluster map: A Clustermap is a heatmap that automatically groups similar rows and columns using hierarchical clustering.** 

```python
sns.clustermap(tipscorr.corr())
```
